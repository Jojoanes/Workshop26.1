```python
import numpy as np
temperaturas = np.random.uniform(20, 35, 7).round(1)
dias = ['segunda', 'terça', 'quarta', 'quinta', 'sexta', 'sábado', 'domingo']
print(temperaturas)
print(dias)
```

```python
# 2. PANDAS
import pandas as pd

df = pd.DataFrame({
    "Dia": dias,
    "Temperatura": temperaturas
})

media = df["Temperatura"].mean()
max_temp = df["Temperatura"].max()
min_temp = df["Temperatura"].min()

print(df)
print(f"\nMédia: {media:.2f}°C | Máx: {max_temp}°C | Mín: {min_temp}°C")

```

```python
# 3. MATPLOTLIB
import matplotlib.pyplot as plt

# 🔹 Gráfico 1
plt.figure()
plt.plot(dias, temperaturas, marker='o')
plt.title("Variação da Temperatura ")
plt.xlabel("Dias")
plt.ylabel("Temperatura (°C)")
plt.grid()
plt.show()

```

```python
# 🔹 Gráfico 2
plt.figure()
plt.bar(dias, temperaturas)
plt.title("Temperatura por Dia ")
plt.xlabel("Dias")
plt.ylabel("Temperatura (°C)")
plt.show()

```

```python
# 🔹 Gráfico 3
plt.figure()
plt.bar(dias, temperaturas)
plt.title("Distribuição das Temperaturas")
plt.xlabel("Dias")
plt.ylabel("Temperatura (°C)")
plt.show()
```
